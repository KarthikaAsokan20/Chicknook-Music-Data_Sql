```python
import pandas as pd
import sqlite3
```

```python
#connecting sql server and read data
connect_data =sqlite3.connect("chinook-music-store-data/chinook.db")
cursor = connect_data.cursor()
```

```python
def sq(q):
    return pd.read_sql_query(q, connect_data).rename(columns = lambda x:x.replace(' ','_').capitalize())
```

```python
tables = sq('''select Name,type from sqlite_master where type='table' ''')
tables
```

```python
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()

# Get column details for each table
for table in tables:
    table_name = table[0]
    cursor.execute(f"PRAGMA table_info({table_name});")
    columns = cursor.fetchall()
    print(f"\nSchema for {table_name}:")
    for col in columns:
        print(col)
```

#### List customer name who bought Genre ="Jazz" with track_id   ####

```python
######connected Genre, track,invoiceline,invoice and customer tables 
query1 =sq('''
with new_table1 as(select t.track_id,g.genre_id,g.name from track t left join genre g on t.genre_id=g.genre_id where g.name ='Jazz'),

new_table2 as (select b.*,c.invoice_id,c.customer_id from 
(select a.*,i.invoice_id as invline_invoice_id from new_table1 a join invoice_line i on a.track_id =i.track_id) as b 
left join invoice c on b.invline_invoice_id=c.invoice_id),

new_table3 as (select distinct c.first_name,c.last_name,n.customer_id,n.genre_id,n.name
from new_table2 n join customer c on c.customer_id = n.customer_id)

select * from new_table3 ;''')
query1
```

#### List Top 10 Customers who spent more money on music ####

```python

q1 =sq('''select c.first_name,c.last_name,i.customer_id,sum(i.total) total_money_spent from invoice i 
 join customer c on c.customer_id =i.customer_id group by i.customer_id order by (total_money_spent) desc limit 10; ''')
q1 
```

### Find the most popular genre (based on track count) ######

```python

sq(''' select g.*,count(t.track_id) as n_tracks from genre g left join track t on g.genre_id= t.genre_id group by 1 order by 3 desc;''')

```

#####  Find the track counts by media type ####

```python
sq(''' select m.*,count(t.track_id) as n_tracks from media_type m join track t on m.media_type_id=t.media_type_id group by 1 ''')
```

#### selling genres from country "France" #####

```python
sq('''select distinct c.country, g.name from customer c join invoice i on c.customer_id=i.customer_id
join invoice_line ivl on i.invoice_id =ivl.invoice_line_id
join track t on ivl.track_id = t.track_id
join genre g on t.genre_id = g.genre_id where c.country="France"; ''')
```
